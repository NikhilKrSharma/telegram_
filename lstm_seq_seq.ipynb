```python
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Bidirectional
import numpy as np
```

```python
seq_lenght = 200
vector_dim = 300
lstm_hidden_size_1 = 64
num_hidden_rep_size_1 = 64
lstm_hidden_size_2 = 32
num_hidden_rep_size_2 = 32
lstm_hidden_size_3 = 8
num_output_tokens = 3

inputs_seq = Input(shape=(seq_lenght, vector_dim))

lstm_out_1 = Bidirectional(LSTM(lstm_hidden_size_1, return_sequences=True))(inputs_seq)

dense_out_1 = Dense(num_hidden_rep_size_1, activation='softmax')(lstm_out_1)

lstm_out_2 = Bidirectional(LSTM(lstm_hidden_size_2, return_sequences=True))(dense_out_1)

dense_out_2 = Dense(num_hidden_rep_size_2, activation='softmax')(lstm_out_2)

lstm_out_3 = Bidirectional(LSTM(lstm_hidden_size_3, return_sequences=True))(dense_out_2)

dense_out_3 = Dense(num_output_tokens, activation='softmax')(lstm_out_3)

model = Model(inputs_seq, dense_out_3)

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```
